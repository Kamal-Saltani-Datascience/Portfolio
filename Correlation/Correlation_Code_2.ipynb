## Calculating Monthly Returns
### to apply correlation and analytics to inform investment decisions for portfolio stocks better, we need to work with induced empirical distribution created from Monthly Return.Â¶



```python
# create a function to calculate Monthly Return

def calculate_monthly_returns(data):
    # Convert 'Date' column to datetime format
    data.loc[:, 'Date'] = pd.to_datetime(data['Date'])
    
    # Sort DataFrame by Date (if not already sorted)
    data = data.sort_values('Date').reset_index(drop=True)
    
    # Calculate daily returns
    data['Daily Return'] = data['Close'].pct_change()
    
    # Group by month and calculate compounded monthly return
    monthly_returns = data.set_index('Date').groupby(pd.Grouper(freq='M'))['Daily Return'].apply(lambda x: (1 + x).prod() - 1) * 100
    
    return monthly_returns

# Sample data (assuming this is already in a DataFrame or can be read into one)


# Calculate and print monthly returns using the function
monthly_returns_NVIDIA = calculate_monthly_returns(NVIDIA[['Date', 'Close']]).values
print(monthly_returns_NVIDIA)

monthly_returns_DJT = calculate_monthly_returns(DJT[['Date', 'Close']]).values
print(monthly_returns_DJT)

monthly_returns_GPRO = calculate_monthly_returns(GPRO[['Date', 'Close']]).values
print(monthly_returns_GPRO)

monthly_returns_AAPL= calculate_monthly_returns(AAPL[['Date', 'Close']]).values
print(monthly_returns_AAPL)
```

## Before applying probability distribution to average data sets for good returns, we should find the right statistics to answer which stock has a better return.

### Which statistics are relevant to investing money and getting a good return ( good investment plan)?
We can consider the expected return or means to find a better return.

### Which statistics can capture Risk?

We can consider the standard deviation of the random return of each one of the stocks to capture the Risk.
## If we think of Monthly Return of each stock as a random variable, then we have:

#### -> Expected Return:  mean
#### -> Risk:             Standard Deviation

```python
summary = pd.DataFrame(Monthly_Return_1[['mean','std']], index=['NVIDIA', 'DJT', 'GPRO', 'AAPL'])

# Calculate the coefficient of variation
summary['Coefficient of Variation '] = (summary['std'] / summary['mean']) 

# Rename columns for clarity
summary = pd.DataFrame(summary.rename(columns={'mean': 'Expected Return (mean) %', 'std': 'Risk (std) %'}))

# Display the updated DataFrame
summary.T.round(2)
```

## What we are interested in is the trade-off between Expected Return and Risk
### Considering the four stocks:

- DJT has higher return with high-risk
- NVIDIA has a lower return with lower risk
- GPRO and AAPL yield small returns and thus can not be considered good stocks.
- the coefficient of Variation captures the Relative Risk: std/mean
